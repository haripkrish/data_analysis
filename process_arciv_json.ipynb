```python
# Find abbreviations with 5 or more letters in abstract
```

```python
import re


def get_abbriviations(line):
    result = re.search(r"\(([A-Za-z][^_/\\<>]{5,})\)", line)
    if result:
        return result.group(1)
```

```python
rdd.filter(lambda x: get_abbriviations(x['abstract'])).count()
```

```python
# Get number of archive method per month
import datetime

def extract_date(date_in):
    d = datetime.datetime.strptime(date_in, "%Y-%m-%d")
    return d.month

extract_date('2023-12-12')
```

```python
rdd.map(lambda x: (extract_date(x['update_date']), 1)).reduceByKey(lambda x, y: x+y).sortBy(lambda l: l[1]).collect()
```

```python
# Average number of pages
```

```python
def get_page(line):
    search = re.findall('\d+ pages', line)
    if search:
        return int(search[0].split(" ")[0])
    else:
        return 0
```

```python
rdd_average = rdd.map(lambda x: get_page(x['comments'] if x['comments'] != None else "None"))

# remove 0:
rdd_average = rdd_average.filter(lambda x: x != 0)

average_counter = rdd_average.count()
average_sum = rdd_average.reduce(lambda x,y: x + y)

average_counter, average_sum
```
